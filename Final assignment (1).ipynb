```python
%%sql 
SELECT COUNT(*) AS TOTAL_CRIMES 
FROM CHICAGO_CRIME_DATA
```

```python
%%sql 
SELECT case_number 
FROM CHICAGO_CRIME_DATA 
WHERE description Like 'M%'
```

```python
%%sql
SELECT COMMUNITY_AREA_NAME, PER_CAPITA_INCOME
FROM CENSUS_DATA
WHERE PER_CAPITA_INCOME < 11000;
```

```python
%%sql 
SELECT CASE_NUMBER, DESCRIPTION
FROM CHICAGO_CRIME_DATA
WHERE DESCRIPTION LIKE '%MINOR%'
```

```python
%%sql
SELECT CASE_NUMBER, PRIMARY_TYPE, DESCRIPTION
FROM CHICAGO_CRIME_DATA
WHERE PRIMARY_TYPE= 'KIDNAPPING' ;
```

```python
%%sql
SELECT DISTINCT (PRIMARY_TYPE)
FROM CHICAGO_CRIME_DATA
WHERE LOCATION_DESCRIPTION LIKE '%SCHOOL%'
```

```python
%%sql
SELECT "Elementary, Middle, or High School" ,AVG(SAFETY_SCORE) AS AVERAGE_SAFETY_SCORE
FROM CHICAGO_PUBLIC_SCHOOLS
GROUP BY "Elementary, Middle, or High School" ;
```

```python
%%sql
SELECT COMMUNITY_AREA_NAME, PERCENT_HOUSEHOLDS_BELOW_POVERTY
FROM CENSUS_DATA
ORDER BY PERCENT_HOUSEHOLDS_BELOW_POVERTY DESC
LIMIT 5 ;
```

```python
%%sql
SELECT COMMUNITY_AREA_NAME
FROM CENSUS_DATA
WHERE HARDSHIP_INDEX = (SELECT MAX(HARDSHIP_INDEX) FROM CENSUS_DATA) ;
```

```python
%%sql 
SELECT community_area_name 
FROM CENSUS_DATA 
WHERE community_area_number = (select A.community_area_number FROM CHICAGO_CRIME_DATA AS A 
GROUP BY A.community_area_number 
ORDER BY COUNT (A.community_area_number) DESC LIMIT 1) ;
```

```python
%%sql
SELECT CCD.COMMUNITY_AREA_NUMBER ,COUNT(CCD.COMMUNITY_AREA_NUMBER) AS NUMBER
FROM CHICAGO_CRIME_DATA AS CCD 
GROUP BY CCD.COMMUNITY_AREA_NUMBER
ORDER BY COUNT(CCD.COMMUNITY_AREA_NUMBER) DESC
LIMIT 1;
```
